```python
# importação de Bibliotecas
import tweepy
import sys
import pandas as pd
```

```python
# Autenticação com as chaves do Twitter como desenvolvedor
#substituir pela leitura do arquivo com as chaves
#####Copiar as chaves que estão no drive
consumer_key = ''
consumer_secret = ''
access_token = ''
access_token_secret = ''
```

```python
#executa a autenticação através da biblioteca Tweepy
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)

api = tweepy.API(auth)
```

```python
#fazer a coleta por hashtag
#keyword = ('#bolsonaro','#lula')
#Fazer a coleta por usuário
#usuario=('@bolsonaro')
# Definição da palavra que será pesquisada no Twitter
keyword = ('bolsonaro or #bolsonaro','lula')
# Busca por palavras-chave
tweets = api.search(q = keyword, count = 10, result_types = 'mixed')

# por default, "count = 100". O valor máximo de count é 18000
# "Lang = pt" => utilizado quando eu quiser somente tweets em português
```

```python
for tweet in tweets:
    print(tweet)
```

```python
#cria estrutura do dicionário e atribui as colunas
tweets_dict = {}
tweets_dict.clear()
tweets_dict = tweets_dict.fromkeys(['created_at', 'id', 'id_str', 'text', 'truncated', 'entities', 'metadata', 'source', 'in_reply_to_status_id', 'in_reply_to_status_id_str', 'in_reply_to_user_id', 'in_reply_to_user_id_str', 'in_reply_to_screen_name', 'user', 'geo', 'coordinates', 'place', 'contributors', 'retweeted_status', 'is_quote_status', 'retweet_count', 'favorite_count', 'favorited', 'retweeted', 'possibly_sensitive', 'lang'])
```

```python
tweets_dict
```

```python
#Limpar o dicionário antes
#this way
#insere o conteudo da variável tweets no dicionário
for tweet in tweets:
    for key in tweets_dict.keys():
        try:
            twvalue = tweet._json[key]
            tweets_dict[key].append(twvalue)
        except KeyError:
            twvalue = ""
            if(tweets_dict[key] is None):
                tweets_dict[key] = [twvalue]
            else:
                tweets_dict[key].append(twvalue)
        except:
            tweets_dict[key] = [twvalue]
```

```python
dfTweets = pd.DataFrame.from_dict(tweets_dict)
```
